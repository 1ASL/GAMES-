```python
import pygame
wid , height = 800,600
ball_wid , ball_height = 6 ,6 
striker_wid , striker_height = 10,70
fps = 60
vel = 10
ball_vel = 5

win = pygame.display.set_mode((wid,height))
black = (0,0,0)
white = (255,255,255)

out_right = pygame.USEREVENT + 1
out_left = pygame.USEREVENT + 2
out_up = pygame.USEREVENT + 3
out_down = pygame.USEREVENT + 4
striker_coll_right = pygame.USEREVENT + 5
striker_coll_left = pygame.USEREVENT + 6

def draw_window(striker_left , striker_right , ball ) :
    win.fill(black)
    pygame.draw.rect(win,white,striker_left)
    pygame.draw.rect(win,white,striker_right)
    pygame.draw.rect(win , white , ball)
    pygame.display.update()



def left_player_movement(key_pressed,striker_left):
    if key_pressed[pygame.K_w] and striker_left.y >= 20:
        striker_left.y -= vel
    if key_pressed[pygame.K_s] and striker_left.y <= height-striker_height :
        striker_left.y += vel

def right_player_movement(key_pressed,striker_right):
    if key_pressed[pygame.K_UP] and striker_right.y >= 20:
        striker_right.y -= vel
    if key_pressed[pygame.K_DOWN] and striker_right.y <= height-70 :
        striker_right.y += vel

def ball_movement(ball , striker_right ,striker_left, ball_vel_x ,ball_vel_y):
    #ball_vel_x = 2
    #ball_vel_y = -2

    if ball.x > wid :
        pygame.event.post(pygame.event.Event(out_right))
    
    if ball.y < 0 :
        pygame.event.post(pygame.event.Event(out_up))
    
    if ball.colliderect(striker_right) :
        pygame.event.post(pygame.event.Event(striker_coll_right))
    
    if ball.x < 0 :
        pygame.event.post(pygame.event.Event(out_left))
    
    if ball.y > height- 10:
        pygame.event.post(pygame.event.Event(out_down))
    
    if ball.colliderect(striker_left) :
        pygame.event.post(pygame.event.Event(striker_coll_left))
    
        
        
        #pygame.display.update()
        #print(ball_vel) 
    
    #print(ball_vel)
    ball.x += ball_vel_x
    ball.y += ball_vel_y
    #print(ball_vel_x)
    


def main():
     
     clock = pygame.time.Clock()

     striker_left = pygame.Rect(5,200,striker_wid,striker_height)
     striker_right = pygame.Rect(wid-20,200,striker_wid,striker_height)
     ball = pygame.Rect(wid//2 , height//2 , ball_wid , ball_height)
     ball_initial = pygame.Rect(wid//2 , height//2 , ball_wid , ball_height)
     #dummy = 0 
     ball_vel_x = ball_vel
     ball_vel_y = -ball_vel
     while True:
        clock.tick(fps)
        
        for events in pygame.event.get():
            if events.type == pygame.QUIT :
                pygame.quit()
            
            if events.type == out_right:
                ball = pygame.Rect(wid//2 , height//2 , ball_wid , ball_height)
                ball_vel_x = ball_vel
                ball_vel_y = -ball_vel
            
            if events.type == out_up :
                ball_vel_y = ball_vel 
                
            if events.type ==  striker_coll_right :   
                ball_vel_x = -ball_vel
            
            if events.type == out_left:
                ball = pygame.Rect(wid//2 , height//2 , ball_wid , ball_height)
                ball_vel_x = ball_vel
                ball_vel_y = -ball_vel

            if events.type == out_down :
                ball_vel_y = -ball_vel 
                
            if events.type ==  striker_coll_left :   
                ball_vel_x = ball_vel
                
                
        key_pressed = pygame.key.get_pressed()
       # print (ball.y)
        right_player_movement(key_pressed,striker_right)
        left_player_movement(key_pressed,striker_left)
        ball_movement(ball , striker_right,striker_left, ball_vel_x , ball_vel_y)
        draw_window(striker_left , striker_right , ball )
        #dummy += 1
        
        

main()

```
